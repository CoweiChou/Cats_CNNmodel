```python
import torch
from torch import nn
import torchvision
from torchvision import models, transforms
from torchsummary import summary
from IPython.display import display

import matplotlib.pyplot as plt
from PIL import Image
import os

import time  # 導入 time 模塊
import torch.nn.functional as F

# 定義圖片預處理操作
preprocess = transforms.Compose([
    transforms.Resize((224, 224)),  # 調整圖片大小
    transforms.ToTensor(),  # 將圖片轉為 Tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # 正規化圖片
])

# 設置測試資料夾路徑
test_dir = "/content/test"  # 替換為你的測試資料夾路徑

# 獲取測試資料夾中的所有圖片文件名稱
image_names = os.listdir(test_dir)

# 載入模型
model = torch.load('model.pth', map_location=torch.device('cpu'))  # 替換為你的模型路徑

# 將模型設置為評估模式
model.eval()

# 將輸入 Tensor 移動到相應的裝置（如 GPU 或 CPU）
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def getModelSize(model):
    param_size = 0
    param_sum = 0
    for param in model.parameters():
        param_size += param.nelement() * param.element_size()
        param_sum += param.nelement()
    buffer_size = 0
    buffer_sum = 0
    for buffer in model.buffers():
        buffer_size += buffer.nelement() * buffer.element_size()
        buffer_sum += buffer.nelement()
    all_size = (param_size + buffer_size) / 1024 / 1024
    print('模型大小： {:.3f}MB'.format(all_size))
    print()
    return (param_size, param_sum, buffer_size, buffer_sum, all_size)

getModelSize(model)

total_time = 0.0
total_score = 0.0

# 遍歷測試資料夾中的所有圖片文件名稱
for image_name in image_names:
    # 組合圖片文件的完整路徑
    image_path = os.path.join(test_dir, image_name)

    # 讀取圖片文件
    image = Image.open(image_path)

    # 預處理圖片
    input_tensor = preprocess(image)
    input_batch = input_tensor.unsqueeze(0)
    input_batch = input_batch.to(device)

    # 記錄開始時間
    start_time = time.time()

    # 對圖片進行預測
    with torch.no_grad():
        output = model(input_batch)

    # 記錄結束時間
    end_time = time.time()

    # 累計運算時間和預測分數
    total_time += end_time - start_time

    # 獲取預測結果
    _, predicted_idx = torch.max(output, 1)
    predicted_class = predicted_idx.item()


    # 計算輸出的概率分佈
    probabilities = F.softmax(output, dim=1)

    # 獲取最大概率對應的類別和概率值
    predicted_probability, predicted_idx = torch.max(probabilities, 1)
    predicted_class = predicted_idx.item()
    predicted_score = predicted_probability.item()

    total_score += predicted_score

    # 處理預測結果(模型訓練時欄位順序如下)
    class_labels =['黑貓', '白貓', '橘貓', '黑白貓', '三花貓', '虎斑貓','玳瑁貓']
    predicted_label = class_labels[predicted_class]

    # 輸出預測結果和運行時間
    print("預測結果：", predicted_label)
    print("運行時間：", end_time - start_time)

    # 輸出預測結果和預測分數
    print("預測分數：", predicted_score)

    # 顯示圖片
    plt.imshow(image)
    plt.show()
    print()

# 計算平均運算速度和平均預測分數
avg_time = total_time / len(image_names)
avg_score = total_score / len(image_names)

# 輸出平均運算速度和平均預測分數
print()
print("平均運算速度：", avg_time)
print("平均預測分數：", avg_score)
```
